```python
print("plot on the left: original grid, on the right: grid cells of size 0.015")
i=0
for k,key in enumerate(allresults.keys()):
    for prange in allresults[key]:
        allpoints=prange[1]
        parameters=allpoints["parameters"].values
        parameters=np.vstack(parameters)
        min_p=parameters.min()
        max_p=parameters.max()
        if min_p<1e-5 or max_p>1e5:
            continue
            i+=1
        
        
        col=allpoints["col"].values
        row=allpoints["row"].values
        #now fill a grid
        pos_ar=np.arange(0.3,2,0.005)
        stp_ar=np.arange(0.2,2,0.005)
        ar=np.zeros((len(stp_ar),len(pos_ar)))
        pos_ar2=np.arange(0.3,2,0.015)
        stp_ar2=np.arange(0.2,2,0.015)
        ar2=np.zeros((len(stp_ar2),len(pos_ar2)))
        for k in range(len(col)):
            p=col[k]
            s=row[k]
            idx_p=np.where(p>pos_ar)[0][-1]
            idx_s=np.where(s>stp_ar)[0][-1]
            ar[idx_s,idx_p]=1
            
            idx_p2=np.where(p>pos_ar2)[0][-1]
            idx_s2=np.where(s>stp_ar2)[0][-1]
            ar2[idx_s2,idx_p2]=1
        fig,axes=plt.subplots(1,2,figsize=(10,4))
        ax=axes[0]
        ax.imshow(ar,origin="lower",cmap=plt.cm.Greys,extent=[pos_ar[0],pos_ar[-1],stp_ar[0],stp_ar[-1]])
        ax.set_title(key+"\n"+"[%2.5f,%2.5f]"%(min_p,max_p))
        ax=axes[1]
        ax.imshow(ar2,origin="lower",cmap=plt.cm.Greys,extent=[pos_ar[0],pos_ar[-1],stp_ar[0],stp_ar[-1]])
        ax.set_title(key+"\n"+"[%2.5f,%2.5f]"%(min_p,max_p))
        for ax in axes:
        #ax.scatter(allpoints["col"],allpoints["row"],s=1,marker="s")
            ax.set_xlabel("position")
            ax.set_ylabel("steepness")
        plt.tight_layout()
        plt.savefig("allpoints_%d.pdf"%i,bbox_inches="tight")
        plt.show()
        i+=1
        
        
```
