```python
import numpy as np
import sys,os
sys.path.append("../bin")
import matplotlib.pyplot as plt
import N6univ
```

```python
instance=N6univ.GRFCalculations_100_100_15()
```

```python
import matplotlib as mpl
mpl.rcParams["font.size"]=22
```

```python
d0=10
n0=0.5*d0
d1=100
n1=0.5*d1
d2=100
n2=0.001*d2
d3=10
n3=1*d3
d4=0.01
n4=1*d4
d5=1
n5=1*d5
d6=0.01
n6=0.1*d6
den=[d0,d1,d2,d3,d4,d5,d6]
num=[n0,n1,n2,n3,n4,n5,n6]
print("\\alpha_0=%g, \\alpha_1=%g, \\alpha_2=%g, \\alpha_3=%g, \\alpha_4=%g, \\alpha_5=%g, \\alpha_6=%g"%tuple(num))
print("\\beta_0=%g, \\beta_1=%g, \\beta_2=%g, \\beta_3=%g, \\beta_4=%g, \\beta_5=%g, \\beta_6=%g"%tuple(den))
parset=np.zeros(14)
parset[::2]=den
parset[1::2]=num
instance.fill_num_den(parset,np.array([0]))

xvals=np.logspace(-3,5,100000)
log10xvals=np.log10(xvals)
out=np.zeros(len(xvals))
for x_,x in enumerate(xvals):
    out[x_]=instance.interfaceGRF(x)
result=instance.interfaceps(minx0=False,maxx1=False,absder=True)
fig,ax=plt.subplots(1,1,figsize=(5,3))
ax.plot(log10xvals,out,color="mediumblue")
ax.set_yticks([])
ax.set_xticks([])
ax.set_xlabel(r"$x$",usetex=True)
x05=np.log10(result[2])
min_=np.min(out)
max_=np.max(out)
f05=(min_+max_)/2
ax.plot([x05,x05],[0,f05],color="k",linewidth=1)
ax.set_ylim([0,1])
ax.set_xlim(log10xvals[0],log10xvals[-1])
colorg="gray"
ax.axhline(y=min_,color=colorg,linewidth=0.5,zorder=-10)
ax.axhline(y=max_,color=colorg,linewidth=0.5,zorder=-10)
ax.axhline(y=f05,color=colorg,linewidth=0.5,zorder=-10)
ax.text(x05,-0.1,r"$x_{0.5}$",horizontalalignment="center",usetex=True)
ax.text(log10xvals[-1]+0.1,min_,r"$m$",usetex=True)
ax.text(log10xvals[-1]+0.1,max_,r"$M$",usetex=True)
ax.text(log10xvals[-1]+0.1,f05-0.03,r"$\frac{m+M}{2}$",usetex=True,fontsize=26)
ax.text(log10xvals[0]-1.2,f05,r"$r(x)$",usetex=True)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.savefig("../figs/Fig4.pdf",bbox_inches="tight")
plt.savefig("../figs/Fig4.svg",bbox_inches="tight")

plt.show()
print()
```
