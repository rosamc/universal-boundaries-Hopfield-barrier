```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
#Figure to show the regions
boundary=np.load("../boundaryanalysis_notebooks/ashapes/N4univ_allregion_2_2_stitched_2023_5_11.npy")
fig,ax=plt.subplots(1,1,figsize=(4.5,4.5))

centre=Rectangle([0.4,0.3], (2.5-0.4), (2.5-0.3),ec="black", fill=False)
ax.plot(boundary[0], boundary[1],color="gray")
ax.fill(boundary[0],boundary[1],color="gray",alpha=0.2)
ax.add_patch(centre)
ax.text(2,2,"center")

centre_ll=Rectangle((0,0),1.5,1.8,ec="magenta", fill=False)
ax.add_patch(centre_ll)
ax.text(0.1,0.1,"lower left",color="magenta")
#ax.plot(s1[0],s1[1],color="r",linewidth=1)

lr=Rectangle([1.2,0],2.1-1.2,0.4,ec="blue",fill=False)
#ax.plot(s2[0],s2[1],color="k",linewidth=3)
ax.add_patch(lr)
ax.text(1.5,0.1,"lower right",color="blue")

ul=Rectangle([0,0.8],0.5,2.1-0.8,ec="green",fill=False)
#ax.plot(s4[0],s4[1],color="k",linewidth=3)
ax.add_patch(ul)
ax.text(0.1,1.5,"upper left",rotation=90,color="green")

ax.set_xlabel("position")
ax.set_ylabel("steepness")



ax.set_title("$\Omega_4$",usetex=True)
plt.savefig("../figs/omega_4_boxes.pdf",bbox_inches="tight")
plt.show()
```
